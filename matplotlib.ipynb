```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 10, 10000)
```

```python
y = np.sin(x) + 2
```

```python
plt.plot(x, y)
plt.savefig("sin.pdf", dpi=100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid(True)
ax.set(xlabel="x", ylabel="sin(x)+2")
fig.savefig("sin.png", dpi=100)
```

```python
norm = np.random.randn(10000) * 5 + 10
plt.hist(norm, 20)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("randn(10000) * 5 + 10")
```

```python
sizes = [10, 20, 15, 5]
labels = ["sing", "dance", "rap", "basketball"]
explodes = [0, 0, 0.1, 0]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, explode=explodes)
ax.axis("equal")
fig.savefig("skill.png", dpi=50)
```

```python
x = np.arange(0, 10, 0.01)
y = np.sin(x) + 2
fig, ax = plt.subplots(2, 2)
# ax[0][0] ax[0][1] ax[1][0] ax[1][1]
ax[0][0].plot(x, x)
ax[0][1].semilogx(x, y)
ax[1][0].semilogy(x, x)
ax[1][1].loglog(x, y)
```
